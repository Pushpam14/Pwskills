```python
#Q4.
"""
The LookupError class is a built-in Python exception class that serves as a base class for exceptions raised when
a key or index lookup fails. It is the parent class of KeyError and IndexError.

Here are explanations of KeyError and IndexError, two common exceptions that inherit from LookupError:

1. KeyError: This error is raised when a dictionary key is not found in the dictionary. For example:

my_dict = {"a": 1, "b": 2, "c": 3}
value = my_dict["d"]

In this example, we try to access a key "d" in my_dict that doesn't exist, causing a KeyError.

2. IndexError: This error is raised when an index is out of range for a list, tuple, or other sequence. For example:

my_list = [1, 2, 3]
value = my_list[3]

In this example, we try to access an index 3 in my_list, which is out of range because my_list only has three elements.

Both of these errors are examples of lookup failures that can occur when trying to access an element by key or index,
and are therefore classified as subclasses of the LookupError class.
"""
```
